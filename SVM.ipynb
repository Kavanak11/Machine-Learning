```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('loan_data_set.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#remove rows that having missing values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.replace({'Loan_Status':{'Y':1,'N':0}},inplace=True)
```

```python
df.Loan_Status
```

```python
df.replace({'Gender':{'Male':1,'Female':0}},inplace=True)
```

```python
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.replace({'Married':{'Yes':1,'No':0},
            'Education':{'Graduate':1,'Not Graduate':0},
            'Self_Employed':{'Yes':1,'No':0}},inplace=True)
df.replace({'Property_Area':{'Urban':1,'Rural':0,'Semiurban':2}},inplace=True)
```

```python
print(df['Dependents'].value_counts())
df.replace({'Dependents':{'3+':4}},inplace = True)
print(df['Dependents'].value_counts())
```

```python
df.head()
```

```python
df.drop('Loan_ID',axis = 1)
```

```python
X = df.drop(columns = ['Loan_ID','Loan_Status'])
X.head()
```

```python
Y = df['Loan_Status']
Y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2,stratify=Y,random_state=34)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn import svm
model = svm.SVC(kernel='linear')
model
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
x_train_pred = model.predict(x_train)
training_acc = accuracy_score(x_train_pred,y_train)
print(f"Training Data accuracy:{training_acc:.2f}")
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_pred)
acc
```

```python
x_test_pred = model.predict(x_test)
testing_acc = accuracy_score(x_test_pred,y_test)
print(f"Testing Data accuracy:{testing_acc:.2f}")
```
