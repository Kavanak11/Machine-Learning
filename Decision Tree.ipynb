DECISION TREE

```python
import pandas as pd
tp = pd.read_csv("C:/Users/Kavanak/Downloads/play_tennis.csv")
tp
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
tp = tp.apply(le.fit_transform)
tp
```

```python
X = tp.drop('PlayTennis',axis = 1)
Y = tp['PlayTennis']
```

```python
X.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.1,random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accu = accuracy_score(y_pred,y_test)
accu
```

```python
import pandas as pd
df = pd.read_csv("C:/Users/Kavanak/Downloads/heart.csv")
df
```

```python
X = df.drop('target',axis = 1)
Y = df['target']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,f1_score
accu = accuracy_score(y_pred,y_test)
accu
```

```python
f1_sc = f1_score(y_test,y_pred)
f1_sc
```

# logistic Regression vs Decision Tree

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 100,random_state = 42)
lr.fit(x_train,y_train)
```

```python
model_pred = lr.predict(x_test)
model_pred
```

```python
from sklearn.metrics import accuracy_score,f1_score
acc = accuracy_score(y_test,model_pred)
acc
```

```python
f1 = f1_score(y_test,model_pred)
f1
```

```python
model_eval = pd.DataFrame({'Model':['LogisticRegr','DecisionTree'],
                           'Accuracies':[acc,accu],
                            'f1_score':[f1,f1_sc]})
model_eval
```
