```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Category'] = le.fit_transform(df['Category'])
```

```python
df['Category']
```

```python
X = df['Message']
Y = df['Category']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2, random_state = 3)
```

```python
X.shape,x_train.shape,x_test.shape
```

```python
Y.shape,y_train.shape,y_test.shape
```

```python
print(x_train.head(2))
```

```python
print(y_train.head(2))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer(stop_words='english',lowercase = True)
x_train_features = vector.fit_transform(x_train)
```

```python
x_test_features = vector.transform(x_test)    # for converting test data, only transform not fit
print(x_test_features)
```

```python
x_test = pd.DataFrame(
    x_train_features.toarray(),
    columns = vector.get_feature_names_out()
)
print(x_test.head())
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_features,y_train)
```

```python
model_pred_train = lr.predict(x_train_features)
model_pred_train
```

```python
from sklearn.metrics import accuracy_score
training_acc = accuracy_score(y_train,model_pred_train)
print("Training Accuracy:",training_acc)
```

```python
model_pred_test = lr.predict(x_test_features)
model_pred_test
```

```python
from sklearn.metrics import accuracy_score
testing_acc = accuracy_score(y_test,model_pred_test)
print("Testing Accuracy:",testing_acc)
```

```python
new_mail = ["Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat..."]

new_mail_f = vector.transform(new_mail)
prediction = lr.predict(new_mail_f)

if prediction[0] == 1:
    print("Spam Mail")
else:
    print("Not Spam Mail")
```
