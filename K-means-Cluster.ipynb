```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv(r"C:\Users\Kavanak\Downloads\income.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Age,df.Salary)
plt.show()
```

```python
km = KMeans(n_clusters=3)
```

```python
#in unsupervised fit_predict used,
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(df[['Age','Salary']])
```

```python
y_pred
```

```python
df['Cluster'] = y_pred
df
```

```python
km.cluster_centers_
```

```python
#separate values into seperate variables

C1 = df.Cluster == 0
print(C1)
```

```python
C1 = df[df['Cluster'] == 0]
C1
```

```python
C2 = df[df['Cluster'] == 1]
C2
```

```python
C3 = df[df['Cluster'] == 2]
C3
```

```python
plt.scatter(C1.Age,C1.Salary,color = 'red')
```

```python
plt.scatter(C1.Age,C1.Salary,color = 'red')
plt.scatter (C2.Age,C2.Salary,color = 'blue')
plt.scatter(C3.Age,C3.Salary,color = 'green')
print(km.cluster_centers_)
plt.xlabel("Ages")
plt.ylabel("Salaries")
print(km.cluster_centers_)
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color = 'purple',marker="*")
```

```python
ms = MinMaxScaler()
ms.fit(df[['Salary']])
df['Salary'] = ms.transform(df[['Salary']])
df
```
